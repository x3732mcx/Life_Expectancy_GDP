# Life Expectancy Project

In this project we will analyze data on GDP (Gross Domestic Product) and life expectancy from the World Heath Organization and the World Bank trying to identify the relationship between the GDP and life expectancy of six countries.

```python
# import libraries
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
from scipy.stats import pearsonr
```

```python
# read and take a look of the data
data = pd.read_csv('all_data.csv')
data.head()
```

```python
# Look for the coorelation
corr_lifeExpec_GDP, p = pearsonr(data['Life expectancy at birth (years)'], data.GDP)
print(corr_lifeExpec_GDP)
```

A correlation larger than 0.3 indicates a linear association, not strong, between Life Expectancy and GDP in these countries.

Let's investigate it.

```python
# Average life expectancy in these nations
countries = data.Country.unique()
print('Life Expectancy Average in these countries.')
for i in range(len(countries)):
    lifeexpect_avg = np.mean(data['Life expectancy at birth (years)'][data.Country == countries[i]])
    print(f'{countries[i]}: {round(lifeexpect_avg, 2)}')
```

```python
print('GDP Average in these countries.')
for j in range(len(countries)):
    gdp_avg = np.mean(data['GDP'][data.Country == countries[j]])
    print(f'{countries[j]}: {gdp_avg}')
```

```python
life_expectancy_avg = np.mean(data['Life expectancy at birth (years)'])
gdp_avg = np.mean(data['GDP'])
```

```python
plt.figure(figsize=(20, 8))
plt.subplot(1,2,1)
sns.boxplot(data=data, x='Country', y='Life expectancy at birth (years)', palette='bright')
plt.axhline(life_expectancy_avg, color='r', label='Average Life Expectancy')
plt.title('Life Expectancy by Country', fontsize=15)
plt.xticks(rotation=30)
plt.legend(loc=6)
plt.grid()

plt.subplot(1,2,2)
sns.boxplot(data=data, x='Country', y='GDP', palette='bright')
plt.axhline(gdp_avg, color='r', label='Average GDP')
plt.title('GDP by Country', fontsize=15)
plt.xticks(rotation=30)
plt.legend(loc=6)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(20, 8))
plt.subplot(1,2,1)
sns.lineplot(data=data, x='Year', y='Life expectancy at birth (years)', hue='Country', palette='bright')
plt.title('Life Expectancy Evolution by Country', fontsize=15)
plt.grid()

plt.subplot(1,2,2)
sns.lineplot(data=data, x='Year', y='GDP', hue='Country', palette='bright')
plt.title('GDP Evolution by Country', fontsize=15)
plt.grid()
plt.show()
```

Here we can see that life the life expectancy has increased over time in these six nations, while GDP has not.

Chile has a very good life expectancy, but a very bad GDP, a bit before Zimbabwe.

The China has increased the life expectancy and it's GDP. 

Germany has the highest life expectancy while Zimbabwe has the lowest.

The United States has the highest GDP while Zimbabwe has the lowest.

Mexico has a good life expectancy, but a bad GDP.
